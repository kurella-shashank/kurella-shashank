```python
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns

def plot_confusion_matrix_and_roc(model, X_test, y_test_encoded, model_name):
    # Confusion Matrix
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test_encoded, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'{model_name} - Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # Classification Report
    print(f'{model_name} - Classification Report')
    print(classification_report(y_test_encoded, y_pred))

    # ROC Curve
    if hasattr(model, "predict_proba"):  # Ensure model supports predict_proba
        y_prob = model.predict_proba(X_test)[:, 1]
        fpr, tpr, thresholds = roc_curve(y_test_encoded, y_prob)
        roc_auc = roc_auc_score(y_test_encoded, y_prob)
        plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], 'k--')
        plt.title(f'{model_name} - ROC Curve')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.legend(loc='lower right')
        plt.show()


```

```python
# Random Forest results
plot_confusion_matrix_and_roc(rf_model, X_test_imputed, y_test_encoded, 'Random Forest')

# Gradient Boosting results
plot_confusion_matrix_and_roc(gb_model, X_test_imputed, y_test_encoded, 'Gradient Boosting')

# XGBoost results
plot_confusion_matrix_and_roc(xgb_model, X_test_imputed, y_test_encoded, 'XGBoost')

# MLP Classifier results
plot_confusion_matrix_and_roc(mlp_model, X_test_imputed, y_test_encoded, 'MLP Classifier')

# LightGBM results
plot_confusion_matrix_and_roc(lgb_model, X_test_imputed, y_test_encoded, 'LightGBM')
```

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind

# Safe T-test function
def perform_t_test_safe(group1, group2):
    """
    Perform T-test safely by handling zero variance, empty groups, and NaN/Inf values.
    """
    if len(group1) == 0 or len(group2) == 0:
        return np.nan, np.nan  # Return NaN if groups are empty
    
    group1 = np.nan_to_num(group1, nan=0.0, posinf=0.0, neginf=0.0)
    group2 = np.nan_to_num(group2, nan=0.0, posinf=0.0, neginf=0.0)
    
    if np.var(group1) == 0 or np.var(group2) == 0:
        return np.nan, np.nan  # Return NaN if variance is zero
    
    return ttest_ind(group1, group2)

# Function to perform T-tests for a specific metric
def perform_t_test(metrics, metric_name):
    """
    Perform T-tests for a specific metric across all model pairs.
    """
    t_test_results = []
    for model1, values1 in metrics.items():
        for model2, values2 in metrics.items():
            if model1 != model2:
                # Extract metric values for the specific metric
                metric_values1 = np.array(values1[metric_name])
                metric_values2 = np.array(values2[metric_name])
                
                # Perform safe T-test
                t_stat, p_value = perform_t_test_safe(metric_values1, metric_values2)
                t_test_results.append({
                    'Metric': metric_name,
                    'Model 1': model1,
                    'Model 2': model2,
                    'T-statistic': t_stat,
                    'P-value': p_value
                })
    
    # Convert results to DataFrame
    t_test_df = pd.DataFrame(t_test_results)
    return t_test_df

# Example usage with metrics dictionary
metrics = {
    'Random Forest': {'Accuracy': [0.9, 0.89, 0.91], 'Precision': [0.88, 0.86, 0.87], 'Recall': [0.9, 0.89, 0.91], 'F1 Score': [0.89, 0.87, 0.88]},
    'Gradient Boosting': {'Accuracy': [0.85, 0.84, 0.86], 'Precision': [0.83, 0.81, 0.82], 'Recall': [0.84, 0.85, 0.83], 'F1 Score': [0.83, 0.82, 0.84]},
    # Add other models' metrics here
}

# Run T-tests for each metric
t_test_accuracy = perform_t_test(metrics, 'Accuracy')
t_test_precision = perform_t_test(metrics, 'Precision')
t_test_recall = perform_t_test(metrics, 'Recall')
t_test_f1 = perform_t_test(metrics, 'F1 Score')

# Display the results for one metric as an example
print("T-Test Results for Accuracy:")
print(t_test_accuracy)

# Save or display other results as needed
print("T-Test Results for Precision:")
print(t_test_precision)

print("T-Test Results for Recall:")
print(t_test_recall)

print("T-Test Results for F1 Score:")
print(t_test_f1)
```

```python
from scipy.stats import ttest_ind
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import StratifiedKFold

# Function to compute metrics for cross-validation
def compute_metrics_for_t_test(model, X, y):
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    accuracies, precisions, recalls, f1_scores = [], [], [], []

    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y[train_index], y[test_index]

        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        accuracies.append(accuracy_score(y_test, y_pred))
        precisions.append(precision_score(y_test, y_pred, average='binary', zero_division=0))
        recalls.append(recall_score(y_test, y_pred, average='binary', zero_division=0))
        f1_scores.append(f1_score(y_test, y_pred, average='binary'))

    return accuracies, precisions, recalls, f1_scores

# Function to perform T-tests
def perform_t_test(metrics_dict, metric_name):
    t_test_results = []
    for model1, values1 in metrics_dict.items():
        for model2, values2 in metrics_dict.items():
            if model1 != model2:
                # Perform T-test
                t_stat, p_value = ttest_ind(values1[metric_name], values2[metric_name], equal_var=False)
                t_test_results.append({
                    'Metric': metric_name,
                    'Model 1': model1,
                    'Model 2': model2,
                    'T-statistic': t_stat,
                    'P-value': p_value
                })

    # Convert results to DataFrame
    t_test_df = pd.DataFrame(t_test_results)
    return t_test_df

# Compute metrics for each model
metrics = {}
models = {
    'Random Forest': rf_model,
    'Gradient Boosting': gb_model,
    'XGBoost': xgb_model,
    'MLP Classifier': mlp_model,
    'LightGBM': lgb_model
}

for model_name, model in models.items():
    acc, prec, rec, f1 = compute_metrics_for_t_test(model, pd.DataFrame(X_train_imputed), y_train_encoded)
    metrics[model_name] = {
        'Accuracy': acc,
        'Precision': prec,
        'Recall': rec,
        'F1 Score': f1
    }

# Run T-tests for each metric
t_test_accuracy = perform_t_test(metrics, 'Accuracy')
t_test_precision = perform_t_test(metrics, 'Precision')
t_test_recall = perform_t_test(metrics, 'Recall')
t_test_f1 = perform_t_test(metrics, 'F1 Score')

# Display T-test results
print("T-Test Results for Accuracy:")
print(t_test_accuracy)

print("T-Test Results for Precision:")
print(t_test_precision)

print("T-Test Results for Recall:")
print(t_test_recall)

print("T-Test Results for F1 Score:")
print(t_test_f1)
```

```python
# Function to perform T-tests
def perform_t_test(metrics_dict, metric_name):
    t_test_results = []
    for model1, values1 in metrics_dict.items():
        for model2, values2 in metrics_dict.items():
            if model1 != model2:
                # Perform T-test
                t_stat, p_value = ttest_ind(values1[metric_name], values2[metric_name], equal_var=False)

                # Format p-value: use scientific notation for very small values
                if p_value < 1e-3:
                    p_value_formatted = f"{p_value:.2e}"  # Scientific notation
                else:
                    p_value_formatted = f"{p_value:.2f}"  # Regular rounded format

                t_test_results.append({
                    'Metric': metric_name,
                    'Model 1': model1,
                    'Model 2': model2,
                    'T-statistic': round(t_stat, 2),
                    'P-value': p_value_formatted
                })

    # Convert results to DataFrame
    t_test_df = pd.DataFrame(t_test_results)
    return t_test_df

# Run T-tests for each metric
t_test_accuracy = perform_t_test(metrics, 'Accuracy')
t_test_precision = perform_t_test(metrics, 'Precision')
t_test_recall = perform_t_test(metrics, 'Recall')
t_test_f1 = perform_t_test(metrics, 'F1 Score')

# Display T-test results
print("T-Test Results for Accuracy:")
print(t_test_accuracy)

print("T-Test Results for Precision:")
print(t_test_precision)

print("T-Test Results for Recall:")
print(t_test_recall)

print("T-Test Results for F1 Score:")
print(t_test_f1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_t_statistics_heatmap(t_test_df, metric_name):
    filtered_df = t_test_df[t_test_df['Metric'] == metric_name]
    heatmap_data = filtered_df.pivot("Model 1", "Model 2", "T-statistic")
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={'label': 'T-statistic'})
    plt.title(f"T-Statistics Heatmap for {metric_name}")
    plt.xlabel("Model 2")
    plt.ylabel("Model 1")
    plt.tight_layout()
    plt.show()
```
